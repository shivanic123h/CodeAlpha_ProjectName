```python
# Features and target
X = df[['year', 'consumption_million_liters']]
y = df['high_production']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

```

```python
# -----------------------------
# Step 1: Create Dataset
# -----------------------------
data = {
    'year': [2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'production_million_liters': [4100, 4200, 4300, 4450, 4600, 4000, 4700],
    'consumption_million_liters': [4000, 4100, 4250, 4400, 4550, 3950, 4650]
}

df = pd.DataFrame(data)

```

```python
# Target: 1 if production > consumption
df['high_production'] = (df['production_million_liters'] > df['consumption_million_liters']).astype(int)

# Sort by year for correct order
df = df.sort_values('year')


```

```python
# -----------------------------
# Step 2: Bar Chart - Production vs Consumption
# -----------------------------
plt.figure(figsize=(10, 6))
bar_width = 0.4
index = range(len(df))

plt.bar([i - bar_width/2 for i in index], df['production_million_liters'], bar_width, label='Production')
plt.bar([i + bar_width/2 for i in index], df['consumption_million_liters'], bar_width, label='Consumption')

plt.xlabel('Year')
plt.ylabel('Million Liters')
plt.title('Coca-Cola Production vs Consumption (Bar Chart)')
plt.xticks(index, df['year'])
plt.legend()
plt.tight_layout()
plt.grid(axis='y')
plt.show()

```

```python
# -----------------------------
# Step 3: Train Decision Tree
# -----------------------------
X = df[['year', 'consumption_million_liters']]
y = df['high_production']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

```

```python
# -----------------------------
# Step 4: Prediction and Evaluation
# -----------------------------
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

```

```python
# -----------------------------
# Step 5: Visualize Decision Tree
# -----------------------------
plt.figure(figsize=(10, 6))
plot_tree(model, feature_names=X.columns, class_names=['Low', 'High'], filled=True)
plt.title("Decision Tree - High Production Prediction")
plt.show()

```

```python
# -----------------------------
# Step 6: Bar Chart - Predicted High Production (1 = Yes, 0 = No)
# -----------------------------
# Predict on full dataset for visualization
df['predicted_high_production'] = model.predict(X)

plt.figure(figsize=(10, 4))
sns.barplot(x='year', y='predicted_high_production', data=df, palette='coolwarm')
plt.title("Predicted: High Production Years (Decision Tree Output)")
plt.xlabel("Year")
plt.ylabel("High Production (1 = Yes, 0 = No)")
plt.ylim(-0.1, 1.1)
plt.grid(True, axis='y')
plt.show()
```
